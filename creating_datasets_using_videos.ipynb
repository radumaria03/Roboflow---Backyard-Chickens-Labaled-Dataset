# Function to extract the frame from the video
 * `frame_interval` - represent how often we want to extract the image (a `30fps video` with `frame_interval=30`, means an image each second)

```python
import cv2
import os


def extract_frames(video_path, output_folder, frame_interval=90):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    video = cv2.VideoCapture(video_path)
    frame_count = 0
    frame_number = 0

    while video.isOpened():
        success, frame = video.read()
        if not success:
            break


        if frame_count % frame_interval == 0:
            frame_number += 1
            output_path = os.path.join(output_folder, f"frame_{frame_number}.jpg")
            cv2.imwrite(output_path, frame)

        frame_count += 1

    video.release()
    print(f"✅ Extracted {frame_number} frames to {output_folder}")
```

```python

extract_frames(f"{video_path}", "chicken_frames_dataset")
```
